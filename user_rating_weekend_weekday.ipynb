```python
import pandas as pd
import numpy as np
import time
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_all = pd.read_csv('data/universal_studio_branches.csv')
data_all
```

```python
days = [datetime.strptime(date, "%B %d, %Y").weekday() for date in data_all['written_date']]
data_all['day'] = days
data_all
```

```python
data_all = data_all.drop_duplicates()
data_all
```

```python
weekend_ratings_all = data_all.loc[(data_all['day'] == 5) | (data_all['day'] == 6)]['rating']
avg_weekend_rating_all = np.mean(weekend_ratings_all)
print('Average weekend rating:', avg_weekend_rating_all)
```

```python
weekday_ratings_all = data_all.loc[(data_all['day'] != 5) & (data_all['day'] != 6)]['rating']
avg_weekday_rating_all = np.mean(weekday_ratings_all)
print('Average weekday rating:', avg_weekday_rating_all)
```

```python
ratings_overall = []
for i in range(7):
    avg_rating = np.mean(data_all.loc[data_all['day'] == i]['rating'])
    ratings_overall.append(avg_rating)
    print(f'day{i}: {avg_rating}')
```

```python
data_after_2018 = pd.read_csv('data/universal_studio_processed_updated.csv')
data_after_2018
```

```python
days = [datetime.strptime(date, "%Y-%m-%d").weekday() for date in data_after_2018['Date']]
data_after_2018['day'] = days
data_after_2018
```

```python
avg_weekend_rating_after_2018 = np.mean(data_after_2018.loc[(data_after_2018['day'] == 5) | (data_after_2018['day'] == 6)]['User_satisfaction'])
print('Average weekend rating:', avg_weekend_rating_after_2018)
```

```python
avg_weekday_rating_after_2018 = np.mean(data_after_2018.loc[(data_after_2018['day'] != 5) & (data_after_2018['day'] != 6)]['User_satisfaction'])
print('Average weekday rating:', avg_weekday_rating_after_2018)
```

```python
ratings_after_2018 = []
for i in range(7):
    avg_rating = np.mean(data_after_2018.loc[data_after_2018['day'] == i]['User_satisfaction'])
    ratings_after_2018.append(avg_rating)
    print(f'day{i}: {avg_rating}')
```

```python
data_all["written_date"] = data_all["written_date"].apply(lambda x: datetime.strptime(x, '%B %d, %Y'))
data_before_2018 = data_all[data_all["written_date"] < datetime.strptime("2018-01-01", '%Y-%m-%d')]
data_before_2018
```

```python
avg_weekend_rating_before_2018 = np.mean(data_before_2018.loc[(data_before_2018['day'] == 5) | (data_before_2018['day'] == 6)]['rating'])
print('Average weekend rating:', avg_weekend_rating_before_2018)
```

```python
avg_weekday_rating_before_2018 = np.mean(data_before_2018.loc[(data_before_2018['day'] != 5) & (data_before_2018['day'] != 6)]['rating'])
print('Average weekday rating:', avg_weekday_rating_before_2018)
```

```python
ratings_before_2018 = []
for i in range(7):
    avg_rating = np.mean(data_before_2018.loc[data_before_2018['day'] == i]['rating'])
    ratings_before_2018.append(avg_rating)
    print(f'day{i}: {avg_rating}')
```

```python
plt.figure(figsize=(10, 6))
X = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
X_axis = np.arange(len(X))
plt.plot(X_axis, ratings_overall, marker='o', label='Overall')
plt.plot(X_axis, ratings_after_2018, marker='o', label='After 2018')
plt.plot(X_axis, ratings_before_2018, marker='o', label='Before 2018')
plt.xticks(X_axis, X)
plt.xlabel("Day of Week")
plt.ylabel("Average Rating")
plt.title("Average Rating vs Day of Week")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plots = sns.barplot(x=X, y=ratings_overall)
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.3f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   xytext=(0, 5),
                   textcoords='offset points')
ax.set(xlabel='Day of Week', ylabel='Average Rating', title='Average Rating vs Day of Week', ylim=(0, 5))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
X = ['Weekday', 'Weekend']
X_axis = np.arange(len(X))
plt.plot(X_axis, [avg_weekday_rating_all, avg_weekend_rating_all], marker='o', label='Overall')
plt.plot(X_axis, [avg_weekday_rating_after_2018, avg_weekend_rating_after_2018], marker='o', label='After 2018')
plt.plot(X_axis, [avg_weekday_rating_before_2018, avg_weekend_rating_before_2018], marker='o', label='Before 2018')
plt.xticks(X_axis, X)
plt.xlabel("Weekday/Weekend")
plt.ylabel("Average Rating")
plt.title("Average Rating vs Weekday/Weekend")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plots = sns.barplot(x=['weekday', 'weekend'], y=[avg_weekday_rating_all, avg_weekend_rating_all])
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.3f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   xytext=(0, 5),
                   textcoords='offset points')
ax.set(ylabel='Average Rating', title='Average Rating on Weekday/Weekend', ylim=(0, 5))
plt.show()
```
